```python
ntable = table[['state','party','votes']]
ntable = ntable['votes'].groupby([ntable['state'],ntable['party']]).sum()
ntable = pd.DataFrame(ntable).reset_index()

ntablesum = ntable.groupby('state').sum().reset_index()
ntablesum.rename(columns={'votes':'sum'}, inplace = True)
ntablemax = ntable.groupby('state').max().reset_index()
ntablemax.rename(columns={'votes':'max'}, inplace = True)
ntablemax['ratio'] = ntablemax['max']/ntablesum['sum']

ntable = ntablemax.join(ntable2, how='left', lsuffix='1',sort=False)
ntable = ntable[['state','party','ratio','POP815213']]
ntable.rename(columns={'POP815213':'avg_multilingual'}, inplace = True)
ntable.sort_values('avg_multilingual')

#ntableR = ntable[(ntable.party == 'Republican')]
#ntableD = ntable[(ntable.party == 'Democrat')]
ntable
```
